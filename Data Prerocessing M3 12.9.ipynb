#### 2.2.1 Handling the genres

Initially, we assessed the prevalence of different genres, identifying 18 key categories as representative:
``'Comedy', 'Drama', 'Romance', 'Action', 'Thriller', 'Crime', 'Documentary', 'Horror', 'Adventure', 'Musical' , 'Animation', 'Science Fiction', 'Fantasy', 'Western', 'War film', 'Sports', 'Hisorty', 'Detective'``.
Subsequent to this, through manual review and data visualization techniques, we refined our genre classifications, integrating some categories in alignment with the genre definitions provided by Wikipedia.

```python
#Statistical word frequency
movies_format['MovieGenres'] = movies_format['MovieGenres'].apply(lambda x: x if isinstance(x, list) else [])
categories = movies_format['MovieGenres'].explode().value_counts()
sorted_category_counts = categories.sort_values(ascending=False)
print(sorted_category_counts.head(50))
```

Among our selected 18 key categories, we first decide to remove movies that contains labels 'animation' or 'documantary'. We remove these movies no matter if the label appears alone or co-occur with other labels. In fact, actors appearing in animated movies mostly partake through voice acting. Given that dubbing an animated action scenes is inherently different from performing in live-action action movies, we decided not to consider animation. Similar reason applies to documantaries.

```python
movies_filtered = movies_format[~movies_format['MovieGenres'].apply(lambda x: any(g in x for g in ['Animation', 'Anime', 'Documentary']))].copy()
```

We then refine movie genres for the rest movies. We only keep genre labels that are in the pre-defined genre list.

```python
#Further cleaning of MovieGenres accoding to our observation

#Common genres of our selection
selected_genres = ['Comedy', 'Drama', 'Romance', 'Action', 'Thriller', 'Crime',\
                   'Horror', 'Adventure', 'Musical', 'Science Fiction', 'Fantasy',\
                   'Western', 'War film', 'Sports', 'Hisorty', 'Detective']

# clean the genre,use deepcopy to make sure there will be no change in previous dataframe
genre = [copy.deepcopy(x) for x in movies_filtered["MovieGenres"].values]
movie_genres_selected = []

#Unify genres‘ data and filter the “drama” genres
for line in genre:
    selected = []
    for g in line:
        if g in selected_genres:
            selected.append(g)
        elif g in ['Romance Film', 'Romantic drama', 'Romantic comedy']:
            selected.append('Romance')
        elif g in ['Crime Fiction', 'Crime Thriller', 'Crime Drama']:
            selected.append('Crime')
        elif g in ['Action/Adventure']:
            selected.append('Adventure')
            selected.append('Action')
        elif g in ['Black comedy', 'Comedy film', 'Comedy-drama', 'Sex comedy', 'Comedy of manners', 'Musical comedy', 'Horror Comedy', 'Fantasy Comedy']:
            selected.append('Comedy')
        elif g in ['Action Thrillers','Wuxia', 'Martial Arts Film']:
            selected.append('Action')
            selected.append('Thriller')
        elif g in ['Detective fiction']:
            selected.append('Detective')
        elif g in ['Historical drama', 'Historical fiction']:
            selected.append('History')
    
    # remove the duplicates and sort list of genres by the initial letter
    sorted_set_genres = sorted(set(selected))
    
    movie_genres_selected.append(sorted_set_genres)
    
movies_genres_filter = movies_filtered.copy()
movies_genres_filter["MovieGenresSelected"] = movie_genres_selected
```

```python
# Inflation per country from 1960 to 2021
df_inflation = pd.read_csv(NEW_DATA_PATH+"inflation.csv", header=2)
df_inflation = df_inflation.iloc[:,0:-1]
df_inflation.head()
```

```python
# Inflation in the USA from 1960 to 2021 as the revenue and budget are calculated in US dollars
usa_inflation = df_inflation[df_inflation['Country Code'] == 'USA']

# Create the inflation coefficients per year 
inflation_coef = np.zeros(63)
inflation_coef[0] = 1
for i in range(62):
    year = i + 1960
    inflation_coef[i+1] = inflation_coef[i] * (1 + float(usa_inflation[str(year)].iloc[0])/100)
print(inflation_coef[-1])
```

```python
movies_revenue_merged_data.info()
```

```python
# Function that gets the coefficient coeff according to the year
def get_inflation(year):
    if isinstance(year, int) :
        x = year-1959
        if x >= 0 and x < len(inflation_coef) :
            return inflation_coef[-1]/inflation_coef[x]
    return float('nan')

# Add the inflation coeff column to the dataframe 
movies_coeff_add = movies_revenue_merged_data.copy()

# Fill NaN values in the MovieReleaseDate column with 0 and convert the data type to integer
movies_coeff_add["MovieReleaseDate"].fillna(0, inplace=True)
movies_coeff_add["MovieReleaseDate"] = movies_coeff_add["MovieReleaseDate"].astype(int)

# Apply the get_inflation function to calculate the inflation coefficient
movies_coeff_add["InflationCoeff"] = movies_coeff_add["MovieReleaseDate"].apply(get_inflation)
```

```python
# Calculate the ‘InflationCorrectedRevenue’
movies_revenue_inflation = movies_coeff_add.copy()
movies_revenue_inflation["InflationCorrectedRevenue"] = movies_revenue_inflation['MovieBoxOfficeRevenue'] * movies_revenue_inflation["MovieReleaseDate"].apply(get_inflation)
```

To investigate the prevalence of missing rating data among actors with varying numbers of film appearances.

```python
# Define the movie count thresholds
thresholds = [5, 10, 15, 20]

# Prepare the results dictionary
results = {'Threshold': [], 'Missing Ratio': []}

# Group the dataframe by 'ActorID' and count the number of movies per actor
movie_count_per_actor = characters_movies_renamed.groupby('ActorID').size()

# Loop over the defined thresholds
for threshold in thresholds:
    # Find actors that meet or exceed the threshold
    actors_meet_threshold = movie_count_per_actor[movie_count_per_actor >= threshold].index
    
    # Filter the dataframe to only include these actors
    filtered_df = characters_movies_renamed[characters_movies_renamed['ActorID'].isin(actors_meet_threshold)]
    
    # Calculate the missing ratio for 'averageRating'
    missing_ratio = filtered_df['AverageRating'].isna().sum() / len(filtered_df)
    
    # Add the results to the results dictionary
    results['Threshold'].append(threshold)
    results['Missing Ratio'].append(missing_ratio)

# Convert the results dictionary to a DataFrame
results_df = pd.DataFrame(results)

# Plot the histogram
plt.bar(results_df['Threshold'], results_df['Missing Ratio'], width=2.5)
plt.xlabel('Number of Movies Threshold')
plt.ylabel('Missing Rating Ratio')
plt.title('Missing Rating Ratio at Different Movie Count Thresholds')
plt.xticks(thresholds)
plt.show()

results_df['Missing Ratio']
```

## 6. Detect change with sliding window

In this section, we use a sliding window to detect changes of genres.

### define functions

```python
def flatten_list(lst):
    """
    Convert a nested list to a flattened list.
    
    Parameters:
    - lst: a nested list whose elements are lists.
    
    Returns:
    - A list. 
    """
    flattened_list = []
    for sublist in lst:
        flattened_list += sublist
        
    return flattened_list


def argmax_list(lst):
    """
    Obtain the most common element(s) in a list.
    
    Parameters:
    - list: a list.
    
    Returns:
    - A list of unique elements with the highest frequency.
    """
    counter_obj = Counter(lst)
    most_common_elements = counter_obj.most_common()
    max_count = most_common_elements[0][1]
    most_common_items = [element for element, count in most_common_elements if count == max_count]
    
    return most_common_items


def check_same_freq(most_common, lst):

    return len(most_common) == len(set(lst))


def check_complete_change(lst1, lst2):
    
    return not bool(set(lst1) & set(lst2))


def jaccard_sim(set1, set2):
    
    return len((set(set1) & set(set2))) / len((set(set1)) | set(set2))


def detect_change(genre_list, hist_size=3, max_sim=0.25):
    
    n_change = 0
    history = []
    change_values = []
    
    # consider all the history
    if hist_size == -1:

        for movie_id, genres in enumerate(genre_list):
            
            if movie_id == 0:
                history.append(genres)
                change_values.append(0)
                continue

            if jaccard_sim(genres, flatten_list(history[:])) <= max_sim and jaccard_sim(genres, flatten_list(history[:])) == 0:
                change_values.append(1)
            else:
                change_values.append(0)
            history.append(genres)

    elif hist_size > 0:
        
        for movie_id, genres in enumerate(genre_list):
            
            if movie_id == 0:
                history.append(genres)
                change_values.append(0)
                continue

            if jaccard_sim(genres, flatten_list(history[-hist_size:])) <= max_sim or jaccard_sim(genres, flatten_list(history[:])) == 0:
                change_values.append(1)
            else:
                change_values.append(0)
            history.append(genres)
    
    return change_values


def extract_bags_of_genres(genre_list, max_window_size=3):
    """
    Extract the bags of genres for each window for a list of genres.
    
    Parameters:
    - genre_list: list of lists, where each sublist contains genres of a movie.
    - window_size: integer, the size of the sliding window.
    
    Returns:
    - A tuple (window_values, argmax_types) where:
        - bags_of_genres: list of the bags_of_genres.
    """
    initial_window_size = int(max_window_size - (max_window_size-1)/2) #This will store initial window_size
    bags_of_genres = []
    
    # Start with the initial window size and increase until we reach the max window size
    for end in range(initial_window_size - 1, len(genre_list)):
        start = max(0, end - max_window_size)  # Determine the start of the window
        window_genres = flatten_list(genre_list[start:end])  # Flatten the genres within the window

        bags_of_genres.append(window_genres)

        # Once we reach the max window size, start moving the start of the window
        if (end-start) == max_window_size:
            start += 1

    # At the end, handle the last windows that are smaller than the max window size
    for window_size in range(max_window_size - 1, initial_window_size - 1, -1):
        start = len(genre_list) - window_size
        window_genres = flatten_list(genre_list[start:])
        
        bags_of_genres.append(window_genres)

    return bags_of_genres


def extract_window_values(genre_list):
    """
    default: window size 3 + window size 5
    """

    bog_3 = extract_bags_of_genres(genre_list, max_window_size=3)
    bog_5 = extract_bags_of_genres(genre_list, max_window_size=5)

    assert len(bog_3) == len(bog_5) == len(genre_list)

    merged_bog = [bog_3[i] + bog_5[i] for i in range(len(genre_list))]
    
    argmax_merged_bog = [argmax_list(bog) for bog in merged_bog]

    return argmax_merged_bog


def count_total_change(genre_list):
    """
    Count the number of total changes, where all genres are new.
    
    Parameters:
    - genre_list: list of lists (length > 1), where each sublist contains genres of a movie.
    
    Returns:
    - An integer that represents the number of total changes.
    """
    n_change = 0
    seen_genres = set()
    
    for movie_id, genres in enumerate(genre_list):
        # if accumulated genre is an empty list
        if movie_id == 0:
            for g in genres:
                seen_genres.add(g)
            continue
         
        curr_n = len(seen_genres)
        for g in genres:
            seen_genres.add(g)
        now_n = len(seen_genres)
        
        if now_n - curr_n == len(genres):
            n_change += 1
    
    return n_change


def calculate_similarity(genre_list):
    
    similarity_list = [1.0]  
    for i in range(1, len(genre_list)):
        set_n_minus_1 = set(genre_list[i - 1])
        set_n = set(genre_list[i])

        similarity_list.append(jaccard_sim(set_n_minus_1, set_n))

    return similarity_list


# Mark interval using numbers
def similarity_intervals_marker(similarity_list):
    intervals = []
    current_interval = []

    for similarity in similarity_list:
        if not current_interval or similarity > 0.5:
            current_interval.append(similarity)
        else:
            intervals.append(current_interval)
            current_interval = [similarity]

    if current_interval:
        intervals.append(current_interval)
        
    distribution = []
#     print(intervals)
    for i, interval in enumerate(intervals):
        distribution.extend([i + 1] * len(interval))

    return distribution


# Merge intervals
def merge_intervals(genre_list, interval_list):
    merged_intervals = []
    for i in range(0, len(interval_list)):
        if i == 0:
            current_genres = set(genre_list[i])
            next_genres = set(genre_list[i + 1])
            if interval_list[i] != interval_list[i + 1]:
                if len(current_genres & next_genres):
                    interval_list[i] = interval_list[i + 1]
                    merged_intervals.append(interval_list[i + 1])
                else:
                    merged_intervals.append(interval_list[i])
            else:
                merged_intervals.append(interval_list[i])
        elif i < len(interval_list) - 1:
            current_genres = set(genre_list[i])
            prev_genres = set(genre_list[i - 1])
            next_genres = set(genre_list[i + 1])
            if interval_list[i] != interval_list[i - 1] and interval_list[i] != interval_list[i + 1]:
                # If current genre and previous genre have intersection, merge into previous interval
                if len(current_genres & prev_genres):
                    interval_list[i] = interval_list[i - 1]
                    merged_intervals.append(interval_list[i - 1])
                # If current genre and next genre have intersection, merge into next interval
                elif len(current_genres & next_genres):
                    interval_list[i] = interval_list[i + 1]
                    merged_intervals.append(interval_list[i + 1])
                # If no intersection, keep it to one interval
                else:
                    merged_intervals.append(interval_list[i])
            else:
                merged_intervals.append(interval_list[i])
        # Handle the last element
        else:
            current_genres = set(genre_list[i])
            prev_genres = set(genre_list[i - 1])
            if interval_list[i] != interval_list[i - 1]:
                if len(current_genres & prev_genres):
                    interval_list[i] = interval_list[i - 1]
                    merged_intervals.append(interval_list[i - 1])
                else:
                    merged_intervals.append(interval_list[i])
            else:
                merged_intervals.append(interval_list[i])
    return merged_intervals


# Tag each interval with word frequency
def tag_intervals(genre_list, new_interval_list):
    tagged_intervals = []

    for interval_number in set(new_interval_list):
        # Get all genres in the current interval
        current_interval_genres = [genre_list[i] for i, interval in enumerate(new_interval_list) if interval == interval_number]
        
        num_elements = sum(1 for interval in new_interval_list if interval == interval_number)
        # Count occurrences of genres
        genre_counts = Counter([genre for sublist in current_interval_genres for genre in sublist])

        # Get the genres with the most occurrences as tags
        most_common_genres = genre_counts.most_common()
        most_common_count = most_common_genres[0][1]
        most_common_genres = [genre for genre, count in most_common_genres if count == most_common_count]
        tagged_intervals.extend([most_common_genres] * num_elements)

    return tagged_intervals


# Genre frequency dictionary
genre_frequency_dict = {
    'Comedy': 19886,
    'Drama': 13173,
    'Romance': 11004,
    'Action': 9921,
    'Thriller': 9438,
    'Crime': 7930,
    'Adventure': 7917,
    'Horror': 5213,
    'Musical': 3935,
    'Science Fiction': 2729,
    'War film': 2627,
    'Fantasy': 2242,
    'Western': 2195,
    'Sports': 1184,
    'History': 899,
    'Detective': 412
}


# Singularize the tag
def singularize_tag(tagged_list, new_interval_list):
    # Combine intersection with previous and next in the first loop
    new_tagged_list = []
    singularize_tagged_list = []
    SN = 0
    for interval_number in set(new_interval_list):
        # Get all markers in the current interval
        num_elements = sum(1 for interval in new_interval_list if interval == interval_number)
        current_tags = tagged_list[SN]
        if len(current_tags) > 1:
            # Find the previous and next interval numbers
            prev_tags = tagged_list[SN - 1] if SN > 0 else []
            next_tags = tagged_list[SN + num_elements] if (SN + num_elements) < len(new_interval_list) else []
            # Take the intersection with the next tags
            if(len(set(current_tags) & set(next_tags))):
                new_tag = set(current_tags) & set(next_tags)
            elif (len(set(current_tags) & set(prev_tags))):
                new_tag = set(current_tags) & set(prev_tags)
            else:
                new_tag = current_tags
            # Add the new tag to the result list
            new_tagged_list.extend([list(new_tag)] * num_elements)
        else:
            # If the tag has only one element, keep it as is
            new_tagged_list.extend([current_tags] * num_elements)
        SN += num_elements  
        
#     print('1st loop:',new_tagged_list)
#     for new_tag in new_tagged_list:
#         print(new_tag)
#     print('--------------------------')
    SN = 0 
    # Single out labels based on frequency in the second loop
    for interval_number in set(new_interval_list):
        # Get all markers in the current interval
        num_elements = sum(1 for interval in new_interval_list if interval == interval_number)
        current_tags = new_tagged_list[SN]
        SN += num_elements  
        if len(current_tags) > 1:
            one_tag = sorted(current_tags, key=lambda x: genre_frequency_dict.get(x, 0), reverse=False)[:1]
            singularize_tagged_list.extend([one_tag] * num_elements)
        else:
            singularize_tagged_list.extend([current_tags] * num_elements)
            
    return singularize_tagged_list


def yield_period_tags(window_value):
    
    similarity_list = calculate_similarity(window_value)
    interval_list = similarity_intervals_marker(similarity_list)
    new_interval_list = merge_intervals(window_value, interval_list)
    tagged_list = tag_intervals(window_value, new_interval_list)
    singularized_tagged_list = singularize_tag(tagged_list, new_interval_list)
    
    return singularized_tagged_list


def detect_change_an_actor(actor_id, df):

    
    df_actor = df[df["ActorID"] == actor_id]
    
    # Calculate the relevant features of the sliding window
    window_value = extract_window_values(df_actor.MovieGenresFinal.tolist())
    period_tag = yield_period_tags(window_value)
    period_change = detect_change(period_tag, hist_size=1, max_sim=0)
    
    df_actor_new = df_actor[["ActorName","ActorID","ActorDOB","ActorHeight","ActorGender","ActorAgeAtMovieRelease",\
                             "Cannes","Berlin","Venice","OscarWinner","OscarNomination","MovieName","MovieReleaseDate",\
                             "MovieRuntime","MovieLanguages","MovieCountries","MovieGenresFinal","AverageRating","NumVotes",\
                             "MovieBoxOfficeRevenue","InflationCoeff","InflationCorrectedRevenue"]].copy()
    
    df_actor_new["WindowValue"] = window_value
    df_actor_new["PeriodChange"] = period_change
    df_actor_new["PeriodTag"] = period_tag
    
    return df_actor_new
```
