```python
import pandas as pd
import json
```

```python
DATA_PATH = "./MovieSummaries/"
movie_columns = ['WID', 'FID', 'MovieName', 'MovieReleaseDate', 'MovieBoxOfficeRevenue', 'MovieRuntime', 'MovieLanguages', 'MovieCountries', 'MovieGenres']
characters_columns = ['WikipediaMovieID','Freebase_MovieID','Movie_release_date','Character_Name','Actor_DOB','Actor_gender','Actor_height','Actor_ethnicity','Actor_Name','Actor_age_at_movie_release','Freebase_character_map1','Freebase_character_map2','Freebase_character_map3']
```

```python
movies = pd.read_csv(DATA_PATH+"movie.metadata.tsv", sep='\t', names=movie_columns)
movies.head()
```

```python
characters = pd.read_csv(DATA_PATH+"character.metadata.tsv", sep='\t', names=characters_columns)
characters.head()
```

```python
#merge two dataframes
characters_movies_merge = pd.merge(characters , movies, how='left', left_on=["WikipediaMovieID", "Freebase_MovieID", "Movie_release_date"], right_on=["WID", "FID", "MovieReleaseDate"])
characters_movies_dropped = characters_movies_merge.drop(["WID", "FID", "MovieReleaseDate"], axis=1).copy()
display(characters_movies_dropped)
display(characters_movies_dropped.columns)
```

```python
# rename each column to snake case for our group
characters_movies = characters_movies_dropped.rename(columns={'WikipediaMovieID': 'wiki_movie_id',
                                                  'Freebase_MovieID': 'freebase_movie_id',
                                                  'Movie_release_date': 'movie_release_date',
                                                  'Character_Name': 'character_name',
                                                  'Actor_DOB': 'actor_dob',
                                                  'Actor_gender': 'actor_gender',
                                                  'Actor_height': 'actor_height',
                                                  'Actor_ethnicity': 'actor_ethnicity',
                                                  'Actor_Name': 'actor_name',
                                                  'Actor_age_at_movie_release': 'actor_age_at_movie_release',
                                                  'Freebase_character_map1': 'freebase_character_actor_map_id',
                                                  'Freebase_character_map2': 'freebase_character_id',
                                                  'Freebase_character_map3': 'freebase_actor_id',
                                                  'MovieName': 'movie_name',
                                                  'MovieBoxOfficeRevenue': 'movie_box_office_revenue',
                                                  'MovieRuntime': 'movie_runtime',
                                                  'MovieLanguages': 'movie_languages',
                                                  'MovieCountries': 'movie_countries',
                                                  'MovieGenres': 'movie_genres'}).copy()

#We've got a merged table of movies and characters
```

```python
# For statistical purposes, convert dictionaries to lists for movie_genres, movie_countries, movie_languages columns
genre_col = []
for l in characters_movies["movie_genres"].values:
    l = json.loads(l)
    genre_col.append(list(l.values()))

country_col = []
for l in characters_movies["movie_countries"].values:
    l = json.loads(l)
    country_col.append(list(l.values()))

language_col = []
for l in characters_movies["movie_languages"].values:
    l = json.loads(l)
    language_col.append(list(l.values()))

characters_movies_astype = characters_movies.copy()
characters_movies_astype["movie_genres"] = genre_col
characters_movies_astype["movie_countries"] = country_col
characters_movies_astype["movie_languages"] = language_col
display(characters_movies_astype)
display(characters_movies_astype.columns)
```

```python
#Read two tables (IMDB ratings and IMDB movies basic info) and merge the desired info together
IMDB_rating = pd.read_csv(DATA_PATH+'title.ratings.tsv', sep='\t')
IMDB_rating.head()
```

```python
IMDB_name = pd.read_csv(DATA_PATH+'title.basics.tsv', sep='\t')
IMDB_name.head()
```

```python
#explore which "title" column to use
differences = IMDB_name[IMDB_name['primaryTitle'] != IMDB_name['originalTitle']]
differences.head()

#Use primaryTitle to merge
```

```python
#Delete unneeded columns
IMDB_name_delete = IMDB_name[['tconst', 'primaryTitle']].copy()

IMDB_merge = IMDB_name_delete.merge(IMDB_rating, on='tconst')
IMDB_merge.isna()['tconst'].unique()
IMDB_merge.head()
```

```python
#Left connection, import movie ratings to our dataframe
characters_movies_IMDBratings_merge = characters_movies_astype.merge(IMDB_merge,left_on='movie_name', right_on='primaryTitle',how = 'left')
characters_movies_IMDBratings_merge.head()
```

```python
# save the dataframe in a binary format
characters_movies_IMDBratings_merge.to_pickle("./characters_movies_rating_.pkl")
```
