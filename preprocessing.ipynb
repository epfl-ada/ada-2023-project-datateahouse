```python
import pandas as pd
import json
```

```python
DATA_PATH = "./MovieSummaries/"
movie_columns = ['WID', 'FID', 'MovieName', 'MovieReleaseDate', 'MovieBoxOfficeRevenue', 'MovieRuntime', 'MovieLanguages', 'MovieCountries', 'MovieGenres']

character_actor_columns = ['WikipediaMovieID','Freebase_MovieID','Movie_release_date','Character_Name','Actor_DOB','Actor_gender','Actor_height','Actor_ethnicity','Actor_Name','Actor_age_at_movie_release','Freebase_character_map1','Freebase_character_map2','Freebase_character_map3']
```

```python
movie_data = pd.read_csv(DATA_PATH+"movie.metadata.tsv", sep='\t', names=movie_columns)
movie_data.head()
```

```python
char_actor_data = pd.read_csv(DATA_PATH+"character.metadata.tsv", sep='\t', names=character_actor_columns)
char_actor_data.head()
```

```python
char_actor_movie_data = pd.merge(char_actor_data, movie_data, how='left', left_on=["WikipediaMovieID", "Freebase_MovieID", "Movie_release_date"], right_on=["WID", "FID", "MovieReleaseDate"]).copy()
char_actor_movie_data_dropped = char_actor_movie_data.drop(["WID", "FID", "MovieReleaseDate"], axis=1).copy()
display(char_actor_movie_data_dropped)
display(char_actor_movie_data_dropped.columns)
```

```python
# rename each column to snake case
char_actor_movie_data_renamed = char_actor_movie_data_dropped.rename(columns={'WikipediaMovieID': 'wiki_movie_id',
                                                  'Freebase_MovieID': 'freebase_movie_id',
                                                  'Movie_release_date': 'movie_release_date',
                                                  'Character_Name': 'character_name',
                                                  'Actor_DOB': 'actor_dob',
                                                  'Actor_gender': 'actor_gender',
                                                  'Actor_height': 'actor_height',
                                                  'Actor_ethnicity': 'actor_ethnicity',
                                                  'Actor_Name': 'actor_name',
                                                  'Actor_age_at_movie_release': 'actor_age_at_movie_release',
                                                  'Freebase_character_map1': 'freebase_character_actor_map_id',
                                                  'Freebase_character_map2': 'freebase_character_id',
                                                  'Freebase_character_map3': 'freebase_actor_id',
                                                  'MovieName': 'movie_name',
                                                  'MovieBoxOfficeRevenue': 'movie_box_office_revenue',
                                                  'MovieRuntime': 'movie_runtime',
                                                  'MovieLanguages': 'movie_languages',
                                                  'MovieCountries': 'movie_countries',
                                                  'MovieGenres': 'movie_genres'}).copy()
```

```python
# convert dictionaries to lists for movie_genres, movie_countries, movie_languages columns

genre_col = []
for l in char_actor_movie_data_renamed["movie_genres"].values:
    l = json.loads(l)
    genre_col.append(list(l.values()))

country_col = []
for l in char_actor_movie_data_renamed["movie_countries"].values:
    l = json.loads(l)
    country_col.append(list(l.values()))

language_col = []
for l in char_actor_movie_data_renamed["movie_languages"].values:
    l = json.loads(l)
    language_col.append(list(l.values()))

char_actor_movie_data_renamed["movie_genres"] = genre_col
char_actor_movie_data_renamed["movie_countries"] = country_col
char_actor_movie_data_renamed["movie_languages"] = language_col
char_actor_movie_data_new = char_actor_movie_data_renamed.copy()
display(char_actor_movie_data_new)
display(char_actor_movie_data_new.columns)
```

```python
# save the dataframe in a binary format
char_actor_movie_data_new.to_pickle("./processed_files/char_actor_movie_data_new.pkl")
```
